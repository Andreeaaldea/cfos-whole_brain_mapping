```python
import os
import pandas as pd
# Example CSV file path
os.chdir("E:\\Andy\\3dplots")
file_path = 'CBTC.csv'

# Read specific columns (e.g., 'Column1' and 'Column3')
columns_to_import = ['Acronym', 'dNormcohen_d']
dd = pd.read_csv(file_path, usecols=columns_to_import)
df = pd.DataFrame(dd)
dict_without_headers = dict(zip(df["Acronym"], df["dNormcohen_d"]))
val_only = df.values
all_areas = df.to_dict()

```

```python
import matplotlib.pyplot as plt

import brainglobe_heatmap as bgh

# Create a list of scenes to plot
# Note: it's important to keep reference to the scenes to avoid a
# segmentation fault
scenes = []
for distance in range(5500, 10500, 500):
    scene = bgh.Heatmap(
        dict_without_headers,
        position=distance,
        orientation="frontal",
        thickness=10,
        format="2D",
        cmap="coolwarm",
        vmax=2,
        vmin=-2,
        label_regions=False,
    )
    scenes.append(scene)

# Create a figure with 6 subplots and plot the scenes
fig, axs = plt.subplots(5, 2, figsize=(20, 20))
for scene, ax in zip(scenes, axs.flatten(), strict=False):
    scene.plot_subplot(fig=fig, ax=ax, show_cbar=True, hide_axes=False)

plt.tight_layout()
plt.show()
```
